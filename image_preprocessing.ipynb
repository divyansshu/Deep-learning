# import necessary libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# load the image


```python
# load the image
# 512 x 512 pixel
img = cv2.imread('/parrot.jpg')

# check if image was loaded successfully
if img is None:
  print("Error: could not load image. please check the file path.")
else:
  print(img.shape)
```

# Matplotlib vs CV2

* openCV read image channels in BGR format by default.
* matplotlib read image channels in RGB format.
* matplotlib cannot display images in BGR format.
* if you display images using matplotlib without converting the image color format to RGB then the color will appear distorted or incorrect.


```python
# import the cv2_imshow function from google.colab.patches
from google.colab.patches import cv2_imshow

cv2_imshow(img)

# cv2.waitKey(0)

# cv2.destroyAllWindows()
```

# convert the BGR color format to RGB color format

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

#display the image

```python
# display the image in RGB color format
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

# displaying image in each color channel

```python
# split the image into its channels

blue, green, red = cv2.split(img_rgb)
```

```python
plt.figure(figsize=(10,10))

# blue channel
plt.subplot(2,2,1)
plt.imshow(blue, cmap='Blues')
plt.title('Blue channel')
# plt.show()

# Green channel
plt.subplot(2,2,2)
plt.imshow(green, cmap='Greens')
plt.title('Green channel')
# plt.show()

# Red channel
plt.subplot(2,2,3)
plt.imshow(red, cmap='Reds')
plt.title('Red channel')

plt.show()
```

# Image Manipulation

```python
# convert an image color to gray
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(8,8))
plt.subplot(2,2,1)

plt.title("GrayScale")
plt.imshow(img_gray, cmap="Grays")
plt.axis('off')
plt.show()
```

# Image Resizing

```python
print(img_rgb.shape)

resized_img = cv2.resize(img_rgb, (100,100))

print(resized_img.shape)
plt.imshow(resized_img)
plt.show()
```
